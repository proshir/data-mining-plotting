```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
data=pd.read_csv('transfusion.data')
data
```

```python
dArray=data.to_numpy()
dArray
```

```python
pd.DataFrame(dArray[:,0])
```

```python
head=['Recency','Frequency','Monetary','Time']
```

```python
for i,name in enumerate(head):
    fig1, ax1 = plt.subplots()
    ax1.set_title(name)
    ax1.boxplot(dArray[:,i])
    plt.savefig(f'Box Plot-{name}.png')
print('\t     Box Plot\n -------------------------------')
```

```python
for i,nameX in enumerate(head):
    for j,nameY in enumerate(head):
        if i==j:
            continue
        fig1, ax1 = plt.subplots()
        x1,y1=dArray[:,i][dArray[:,-1]==0],dArray[:,j][dArray[:,-1]==0]
        ax1.scatter(x1,y1,marker='o',alpha=0.8)
        x2,y2=dArray[:,i][dArray[:,-1]==1],dArray[:,j][dArray[:,-1]==1]
        ax1.scatter(x2,y2,marker='+',alpha=0.8)
        ax1.set_ylabel(nameY, fontsize='medium') 
        ax1.set_xlabel(nameX, fontsize='medium')
        plt.savefig(f'{nameX}-{nameY}-Scatter.png')
print('\t     Scatter Plot\n -----------------------------------')
```
